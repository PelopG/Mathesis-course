```python
import pandas as pd
pd.set_option('display.max_colwidth', 25)
gdp_df = pd.read_csv('worldbank/gdp_per_capita_ppp/API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv', 
                     skiprows=4)
```

```python
gdp_df
```

```python
# We need to remove the last column [ Reminder: this was creater due to comma seperation of the data]

gdp_df.drop(gdp_df.columns[len(gdp_df.columns) - 1], inplace=True, axis=1)
```

```python
gdp_df
```

From the above data we can see that some of the countries correspond to rather a group of countries

```python
# We can find that corresponding lines from the Metadata_County folder
countries_df = pd.read_csv('worldbank/gdp_per_capita_ppp/Metadata_Country_API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_3469401.csv')
countries_df
```

```python
# Drop NaN from the Region Column and the last Column
countries_df.drop(countries_df.columns[len(countries_df.columns) - 1], inplace=True, axis=1)
countries_df = countries_df.loc[countries_df['Region'].notnull()]
countries_df
```

```python
# Merging the dataframes together - getting the Country Code , Region
gdp_countries_df = pd.merge(gdp_df, 
                            countries_df[["Country Code", "Region"]], 
                            how="inner", 
                            on="Country Code")
```

```python
gdp_countries_df
```

```python
# life_expectancy data
lex_df = pd.read_csv('worldbank/life_expectancy/API_SP.DYN.LE00.IN_DS2_en_csv_v2_3469289.csv', 
                     skiprows=4)
```

```python
lex_df
```

```python
# Dropping the last column
lex_df.drop(lex_df.columns[len(lex_df.columns) - 1], inplace=True, axis=1)
```

```python
lex_df
```

```python
# Merge together the 2 dataframes ( life-expectancy and gdp_countries )
# Merge will be done on the Country Code coloumn
lex_gdp_countries_df = pd.merge(gdp_countries_df, 
                                lex_df.loc[:,'Country Code':], 
                                how="inner", 
                                on="Country Code",
                                suffixes=("_gdp", "_lex"))
```

We have to be careful because some of the coloumns in the 2 dataframes have the same name.
If we do not declare a seperation(suffix) then pandas will set to _x & _y

```python
lex_gdp_countries_df
```

```python
# We can now use the above data of 2019 to design a scatter plot 
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10, 8))
_ = plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
```

The above scatter plot shows that the countries are seperated into 2 different categories
1) A group with small life expectancy and small differences in gdp - found on the right hand side of the scatter plot

2) A group with small life expectancy difference and large gdp  >= 25000

```python
# Hence using different colors and taking patches.Rectangle
import matplotlib.patches as patches

plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
rect1 = patches.Rectangle((0, 52), width=25000, height=30, color='blue', alpha=0.2)
rect2 = patches.Rectangle((25000, 72), width=120000, height=14, color='gold', alpha=0.2)
plt.gca().add_patch(rect1)
_ = plt.gca().add_patch(rect2)
```

We can also be intersted in finding a mathematical equation which will explain the correlation between gdp and life-expectancy 

This equation is : y = a + b log(x)

```python
# Hence creating a logarithmic scale scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(x=lex_gdp_countries_df['2019_gdp'], y=lex_gdp_countries_df['2019_lex'])
plt.xscale('log')
```

Eventhough the above graph gives us nice information to get a conclusion from it.

We might also need to consider the population of each country(dot), because comparison for example between india and malta will need to be taken into consideration

```python
# Taking the population data
population_df = pd.read_csv('worldbank/population/API_SP.POP.TOTL_DS2_en_csv_v2_3603750.csv', 
                            skiprows=4)
population_df.drop(population_df.columns[len(population_df.columns) - 1], 
                   inplace=True, axis=1)
```

```python
population_df
```

```python
# We now need to again merge the dataframes together ( gdp & life-expectancy) with population dataframe
pop_lex_gdp_countries_df = pd.merge(lex_gdp_countries_df, 
                                    population_df, 
                                    how="inner", 
                                    on="Country Code")
```

```python
pop_lex_gdp_countries_df
```

```python
# For declaration reasons we rename the columns of population to end with _pop
pop_column_names = { c: f'{c}_pop' for c in population_df.columns if '1960' <= c <= '2020'}
pop_lex_gdp_countries_df.rename(columns=pop_column_names, inplace=True)
```

```python
pop_lex_gdp_countries_df
```

```python
# Ploting now a graph taken into consideration the population dataframe
plt.figure(figsize=(10, 8))
_ = plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], y=pop_lex_gdp_countries_df['2019_lex'], 
                s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
                alpha=0.5)
```

Lets see now if we are able to geographically map the individual countries

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(10, 8))
plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], 
            y=pop_lex_gdp_countries_df['2019_lex'], 
            s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
            alpha=0.5)
plt.xscale("log")

formatter = FuncFormatter(lambda v, p: "%dK" % (v / 1000))

plt.gca().xaxis.set_major_formatter(formatter)
```

```python
# Hence will need to create a categorical variable in ordet to match a number with the area
region_cat = pop_lex_gdp_countries_df['Region'].sort_values().astype('category')
pop_lex_gdp_countries_df['Region Code'] = region_cat.cat.codes
```

```python
pop_lex_gdp_countries_df
```

```python
# Hence we can therefore create a graph
plt.style.use('seaborn-whitegrid')

matplotlib.rcParams.update({'font.size': 16})
plt.figure(figsize=(10, 8))
scatter = plt.scatter(x=pop_lex_gdp_countries_df['2019_gdp'], 
            y=pop_lex_gdp_countries_df['2019_lex'], 
            s=pop_lex_gdp_countries_df['2019_pop']/(2*10**5),
            c=pop_lex_gdp_countries_df['Region Code'],
            alpha=0.6,
            edgecolors='k',
            label=pop_lex_gdp_countries_df['Region'],
            cmap='Spectral_r')

handles, labels = scatter.legend_elements(prop="colors", num=len(region_cat.cat.categories), 
                                          size=10, mec='k', alpha=0.6)
labels = region_cat.cat.categories
plt.legend(handles, labels, loc="lower right")
plt.xscale("log")
formatter = FuncFormatter(lambda v, p: "%dK" % (v / 1000))
plt.gca().xaxis.set_major_formatter(formatter)
```
