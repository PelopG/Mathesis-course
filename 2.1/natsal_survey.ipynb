```python
# We are going to see a new library here ; seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
natsal = pd.read_csv('natsal_het_sex_partners.csv')
natsal
```

```python
# Using describe() we are able to get a better idea of the number of sexual partners
natsal.groupby('Sex').describe()
```

```python
# created a function to for the value_counts
natsal.groupby('Sex').agg(lambda x: x.value_counts().index[0])
```

```python
# First step is to create a graph
sns.set_style('darkgrid')
sns.displot(natsal, x="NumPartners")
```

We can see from the above graph is skew

```python
# In order to avoid this skewness of the data on the left hand side
# Lets focus on the data <50 partners
natsal_50 = natsal[natsal['NumPartners'] <= 50]
sns.displot(natsal_50, x="NumPartners")
```

```python
# Lets now make the above graph a bit larger and more large scale

sns.set_theme(font_scale=1.5)
_ = sns.displot(natsal_50,x="NumPartners", height=7,aspect=1.5) # aspect  = heght to base ratio
```

```python
# Lets see now how many are male or female
_ = sns.displot(natsal_50,x="NumPartners",hue="Sex", height=7,aspect=1.5)
```

```python
# Changing the colors
_ = sns.displot(natsal_50, x="NumPartners",
                hue="Sex",
                palette={'M': 'deepskyblue', 'F': 'pink'},
                height=7, aspect=1.5)
```

```python
# Moving the bars not to be on top of each other
_ = sns.displot(natsal_50, x="NumPartners",
                hue="Sex",
                palette={'M': 'deepskyblue', 'F': 'pink'},
                multiple="dodge", # multiple variable to move the bars
                height=6, aspect=1.5)
```

```python
# Changing the bars outline
_ = sns.displot(natsal_50, x="NumPartners",
                hue="Sex",
                edgecolor='black', palette={'M': 'deepskyblue', 'F': 'pink'}, 
                multiple="dodge",
                height=6, aspect=1.5)
```

```python
# We can also create box-plots
plt.figure(figsize=(16, 8))
_ = sns.boxplot(x=natsal_50['NumPartners'],
                y=natsal_50['Sex'],
                palette={'M': 'deepskyblue', 'F': 'pink'})
```

```python
# We can also create a strip-plot
# This graph advantage is that is can show the population frequency
plt.figure(figsize=(16, 8))
_ = sns.stripplot(x=natsal_50['NumPartners'],
                  y=natsal_50['Sex'],
                  linewidth=0.5, # mark line
                  size=7, # mark size
                  jitter=.25, # spread around the middle
                  edgecolor='black', palette={'M': 'deepskyblue', 'F': 'pink'})
```

```python
# Can also create a violin-plot
plt.figure(figsize=(16, 8))
_ = sns.violinplot(x=natsal_50['NumPartners'],
                   y=natsal_50['Sex'],
                   edgecolor='black',
                   palette={'M': 'deepskyblue', 'F': 'pink'})
```
