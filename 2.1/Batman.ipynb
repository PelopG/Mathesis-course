```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()
```

Pyplot is not the only way.
We have the ability to work with API (Application Programming Interface), where all the elements of the graph are created as intances of a class

```python
# We design the same graph as above but by designing FIGURES and AXES
fig = plt.figure()  # an empty figure with no axes

axes = fig.gca()
axes.plot(x, x, label='linear')

axes.plot(x, x**2, label='quadratic')
axes.plot(x, x**3, label='cubic')

axes.set_xlabel('x label')
axes.set_ylabel('y label')

axes.set_title("Simple Plot")

axes.legend()
```

The general design of a graph using matplotlib can be set with style
matplotlib has a single number set to as pre-configured style

```python
print(plt.style.available)
len(plt.style.available)
```

```python
plt.figure(figsize=(12, 130))
plt.subplots_adjust(hspace=0.6)

x = np.linspace(0, 2, 100)

nstyles = len(plt.style.available)
for i, style in enumerate(plt.style.available):# 26 styles available
    plt.subplot(nstyles, 2, i + 1)
    plt.style.use(style)
    plt.plot(x, x, label='linear')
    plt.plot(x, x**2, label='quadratic')
    plt.plot(x, x**3, label='cubic')

    plt.xlabel('x label')
    plt.ylabel('y label')

    plt.title(style)

    plt.legend()
```

```python
# We can also set diffrent setting for the legend

legend_loc_str = [
    'best',
    'upper right',
    'upper left',
    'lower left',
    'lower right',
    'right',
    'center left',
    'center right',
    'lower center',
    'upper center',
    'center'
]

plt.style.use('ggplot')
plt.figure(figsize=(12, 24))
plt.subplots_adjust(hspace=0.4)

x = np.linspace(0, 2, 100)

for i, loc in enumerate(legend_loc_str):
    plt.subplot(6, 2, i + 1)
    plt.plot(x, x, label='linear')
    plt.plot(x, x**2, label='quadratic')
    plt.plot(x, x**3, label='cubic')

    plt.xlabel('x label')
    plt.ylabel('y label')

    plt.title("Simple Plot, legend at = " + loc)

    plt.legend(loc=loc)
```

```python
# Using matplotlib using figures and axes
figure = plt.figure(figsize=(12, 24))
figure.subplots_adjust(hspace=0.4)

x = np.linspace(0, 2, 100)

for i, loc in enumerate(legend_loc_str):
    ax = figure.add_subplot(6, 2, i + 1)
    ax.plot(x, x, label='linear')
    ax.plot(x, x**2, label='quadratic')
    ax.plot(x, x**3, label='cubic')

    ax.set_xlabel('x label')
    ax.set_ylabel('y label')

    ax.set_title("Simple Plot, legend at = " + loc)

    ax.legend(loc=loc)
```

matplotlib 8 colours:

b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white

```python
colors = [
    'b',
    'g',
    'r',
    'c',
    'm',
    'y',
    'k',
    'w',
]

figure = plt.figure(figsize=(12,24))
figure.subplots_adjust(hspace=0.4)

x = np.linspace(-1, 1, 100)
y = x * np.sin(1/x)

for i,color in enumerate(colors):
    ax = figure.add_subplot(4, 2, i+1)
    # this expression is as such that matplotlib understand that we want a mathematical function
    # r'$y=x \ sin(1/x)$'
    ax.plot(x,y,color=color,label=r'$y=x \ sin(1/x)$') 
    
    ax.set_xlabel('x label')
    ax.set_ylabel('y label')
    
    ax.set_title("Simple Plot, color set at " + color)
    
    ax.legend(loc='best')
```
